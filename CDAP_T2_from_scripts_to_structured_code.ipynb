## 2. **Functions**
A function is a black-box part of the code used to improve **readability** of the main/backbone code.
When Python hits a function call, it doesn't just read the next line. It performs a Control Flow Jump: it "pauses" the main script, it "jumps" to the function block, it creates a Local Scope (a temporary memory space), runs the code and eventually, goes back to the exact spot left carrying the return value like a courier.
Depending on the complexity of the code, you can define the function before in the same source file and directly use it, or you can define it in a different source file and **import** it.

**2.1. Syntax**
``` python
def myFunction(list_of_input):      # Order matters
---> """
     Block note usually exploited to explain the features of the
     function:
     - input:....
     - output:... Force [kN]..
     """
     myVariable = 1   # Valid only inside the function

     # Do something...
     return list_of_output
```

````python
# A function to retrieve the max in between two numbers:

def myMax(a,b):
   if a>b:
     return a
   else:
     return b
a = 50
b = 5
print(myMax(a,b))
````

````python
# If I have to output more than one variables:
def toUpperLower(input_string):
    upper_string = input_string.upper()
    lower_string = input_string.lower()
    return upper_string, lower_string

myString = 'CiaoCIAOciao'
upper, lower = toUpperLower(myString)
print(upper)
print(lower)
uplow = toUpperLower(myString)
print(uplow)  # In this case I will obtain a tuple (Not changeble item)
````

````python
# Parameters of a function:
def myGreater(a,b=0):       # b=0 as default value, if nothing is specified
   return a>b
result = myGreater(-1)
print(result)

# It is possible to reassign the value of b:
result = myGreater(-1,-10)
print(result)
````

````python
# Named parameters:
def myFunction(a,b,c,d):
    print(a)
    print(b)
    print(c)
    print(d)
    return a+b+c+d

sum = myFunction(d=1,a=2,c=3,b=4)  # The order doesn't matter if you explicitely call the input variables
print(sum)

sum_2 = myFunction(1,2,3,4)        # In this case the value are assigned order-wise
print(sum_2)

d=4
sum_3 = myFunction(a=1, b=2, c=3, d=d)
print(sum_3)
````

**2.2. Higher-order function**
In the previous examples we gave variables as input to a function and retrieve output variables. If the input and output of a functions are functions themselves, this is called **higher-order** computing. In the following, it is presented `map` and its syntax. This command will apply a function to all the elements of a list.

````python
# map(function, iterable,...) indicate the list of input to the functions
def toUpper(input_string):
    return input_string.upper()

myList = ['Halo','CIao']
upperList = map(toUpper,myList)
print(upperList)
print(list(upperList))
````

**2.3. Anonymous function**
It is called `lambda`: deep inside Python treats it as a function and it returns a function object. Moreover, it is very useful for simple one-liner function (it avoids to define a function from scratch) and whose result is immediately returned. \
``` python
result = lambda parameters:expression
```

````python
# Convert a unit of measure:

diameter_cm = 20
diameter_inch = lambda x: x/2.54
print(f"The lambda fun. is useful to convert a measure from {diameter_cm} cm to {diameter_inch(diameter_cm)} inches")
````

````python
# Convert a list of measures:

diameter_cm = [20,30,40]
diameter_inch = list(map(lambda x: x/2.54, diameter_cm))   # Need to declare as a list()
print(diameter_inch)
````

```python
class TensileBar:
    # 1. THE CONSTRUCTOR (Identity & State)
    def __init__(self, E, phi):
        self.E = E                  # Attribute: steel Young Modulus in MPa
        self.phi = phi              # Attribute: Cross-section diameter in mm
        self.force = 0              # Initial state: No load applied yet

    # 2. THE METHODS (Behavior)
    def apply_load(self, normal_force):
        """Changes the state of the object."""
        self.force = normal_force
        print(f"Applied {self.force} N to {self.name}")

    def get_area(self):
        """Calculates internal stress based on current state."""
        import math
        area = math.pi * ((self.phi/2)**2)
        return area

     def get_resistance(self,area):
        """Calculates internal stress based on current state."""
        resistance = area * self.E
        return resistance
```

````python
# Part 1


class Product:

  def __init__(self, product_name, category, price, discount_price, quantity, unit_of_measure):
    self.product_name = product_name
    self.category = category
    self.price = price
    self.discount_price = discount_price
    self.quantity = quantity
    self.unit_of_measure = unit_of_measure

  def get_current_selling_price(self):
    if 0 < self.discount_price < self.price:
      return self.discount_price
    return self.price

  def calculate_discount_percentage(self):
    if self.price == 0 or self.discount_price == 0:
        return "0.00%"
    discount = ((self.price - self.discount_price) / self.price) * 100
    return f"{discount:.2f}%"

  def update_stock(self, change):
    if self.quantity + change < 0:
      return False
    self.quantity += change
    return True

  def display_info(self):
      """Prints a formatted summary of the product."""
      selling_price = self.get_current_selling_price()
      discount_pct  = self.calculate_discount_percentage()
      print(
          f"  {'─' * 50}\n"
          f"  Product       : {self.product_name}\n"
          f"  Category      : {self.category}\n"
          f"  Original Price: ${self.price:.2f}\n"
          f"  Discount Price: ${self.discount_price:.2f}\n"
          f"  Selling Price : ${selling_price:.2f}\n"
          f"  Discount      : {discount_pct}\n"
          f"  Stock         : {self.quantity} {self.unit_of_measure}"
      )
````

````python
# Part 2

class GroceryInventoryManager:

    def __init__(self):
        self.inventory = []

    def add_product(self, product_obj):
      for product in self.inventory:
        if product.product_name == product_obj.product_name:
          product.quantity += product_obj.quantity
          return
      self.inventory.append(product_obj)

    def remove_product(self, product_name):
      for product in self.inventory:
          if product.product_name == product_name:
              self.inventory.remove(product)
              return
      print(f"Product '{product_name}' not found in inventory.")

    def find_product(self, product_name):
        for product in self.inventory:
            if product.product_name == product_name:
                return product
        return None

    def get_total_inventory_value(self):
        total = 0.0
        for product in self.inventory:
            total += product.get_current_selling_price() * product.quantity
        return total

    def list_products_by_category(self, category):
        return [product for product in self.inventory if product.category == category]

    def display_all_products(self):
        for product in self.inventory:
            product.display_info()
````

````python
# Part 3

# 1. Initialize Manager
manager = GroceryInventoryManager()

# 2. Populate Inventory from df_as_dict
for item in df_as_dict:
    product = Product(
        product_name=item['ProductName'],
        category=item['Category'],
        price=float(item['Price']),
        discount_price=float(item['DiscountPrice']),
        quantity=float(item['qty']),
        unit_of_measure=item['unit_of_measure']
    )
    manager.add_product(product)

print(f"Inventory populated with {len(manager.inventory)} unique products.\n")

# 3a. Display the first 10 products
print("=" * 55)
print("  FIRST 10 PRODUCTS IN INVENTORY")
print("=" * 55)
for product in manager.inventory[:10]:
    product.display_info()

# 3b. Find and display a specific product
print("\n" + "=" * 55)
print("  FIND: 'Onion (Loose)'")
print("=" * 55)
onion = manager.find_product('Onion (Loose)')
if onion:
    onion.display_info()
else:
    print("Product 'Onion (Loose)' not found in inventory.")

# 3c. Simulate a sale: sell 1.5 kg of 'Onion (Loose)'
print("\n" + "=" * 55)
print("  SALE: Selling 1.5 kg of 'Onion (Loose)'")
print("=" * 55)
if onion:
    success = onion.update_stock(-1.5)
    if success:
        print("Sale of 1.5 kg successful!")
    else:
        print("Sale failed: insufficient stock.")
    onion.display_info()

# 3d. Simulate a restock: add 10.0 kg of 'Onion (Loose)'
print("\n" + "=" * 55)
print("  RESTOCK: Adding 10.0 kg of 'Onion (Loose)'")
print("=" * 55)
if onion:
    onion.update_stock(10.0)
    print("Restocked 10.0 kg successfully!")
    onion.display_info()

# 3e. Attempt to sell more than available stock
print("\n" + "=" * 55)
print("  OVERSELL ATTEMPT on 'Onion (Loose)'")
print("=" * 55)
if onion:
    oversell_qty = onion.quantity + 100
    success = onion.update_stock(-oversell_qty)
    if not success:
        print(
            f"Sale FAILED: tried to sell {oversell_qty} {onion.unit_of_measure} "
            f"but only {onion.quantity} {onion.unit_of_measure} in stock."
        )

# 3f. Total inventory value
print("\n" + "=" * 55)
total_value = manager.get_total_inventory_value()
print(f"  TOTAL INVENTORY VALUE: ${total_value:,.2f}")
print("=" * 55)
````
